```python
metadata_path = "/home/tolis/Desktop/tolis/DNN/project/DeepLearning_2024_2025_DSIT/output/metadata.json"
```

```python
import json
from IPython.display import Image, display


with open(metadata_path, 'r') as f:
    metadata = json.load(f)

# Iterate through each record in the metadata
for record in metadata:
    # Extract the filename and relevant_passages fields
    filename = record.get("filename")
    relevant_passages = record.get("relevant_passages")
    
    if filename and relevant_passages:
        # Construct the full path to the image
        image_path = f"{output_path}/{filename}"
        
        # Display the image
        print(f"Displaying image: {filename}")
        display(Image(filename=image_path))
        
        # Display the relevant passages
        print("Relevant Passages:")
        for i, passage in enumerate(relevant_passages, 1):
            print(f"Passage {i}: {passage}")
    else:
        print("Missing 'filename' or 'relevant_passages' field in record.")
```
